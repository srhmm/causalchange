```python
import pandas as pd

import matplotlib.pyplot as plt
import seaborn as sns
```

```python

coviddf = pd.read_csv("datasets/chapter1_simpson/simpsons_paradox_covid_uk.csv")
 #  accessed here, https://www.openintro.org/data/index.php?data=simpsons_paradox_covid

df = coviddf
cfr_vacc = len(df[(df['vaccine_status']=='vaccinated') & (df['outcome']=='death')] ) / len(df[(df['vaccine_status']=='vaccinated')] )
cfr_unvacc = len(df[(df['vaccine_status']=='unvaccinated') & (df['outcome']=='death')] )/ len(df[(df['vaccine_status']=='unvaccinated')] )

print(f"CFR, vaccinated: {cfr_vacc}\nCFR, unvaccinated: {cfr_unvacc}")
```

```python

cfr_vacc_under = len(df[(df['age_group']=='under 50') & (df['vaccine_status']=='vaccinated') & (df['outcome']=='death')] ) / len(df[(df['age_group']=='under 50') &(df['vaccine_status']=='vaccinated')] )
cfr_unvacc_under = len(df[(df['age_group']=='under 50') &(df['vaccine_status']=='unvaccinated') & (df['outcome']=='death')] )/ len(df[(df['age_group']=='under 50') &(df['vaccine_status']=='unvaccinated')] )

cfr_vacc_over = len(df[(df['age_group']=='50 +') & (df['vaccine_status']=='vaccinated') & (df['outcome']=='death')] ) / len(df[(df['age_group']=='50 +') & (df['vaccine_status']=='vaccinated')] )
cfr_unvacc_over= len(df[(df['age_group']=='50 +') &(df['vaccine_status']=='unvaccinated') & (df['outcome']=='death')] )/ len(df[(df['age_group']=='50 +') & (df['vaccine_status']=='unvaccinated')] )

print(f"Under 50 years old \n\tCFR, vaccinated: {cfr_vacc_under}\n\tCFR, unvaccinated: {cfr_unvacc_under}")
print(f"50+ years old \n\tCFR, vaccinated: {cfr_vacc_over}\n\tCFR, unvaccinated: {cfr_unvacc_over}")
```
